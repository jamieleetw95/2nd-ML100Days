```python
target = app_train["AMT_INCOME_TOTAL"]
print(target)
```

```python
# 計算資料分散程度
# 最小值
print(target.min())

# 最大值
print(target.max())

# 四分位差
print(np.percentile(target, [25, 50, 75]))
print(np.percentile(target,100))

# 變異數
print(target.var())

# 標準差
print(target.std())
```

```python
# 計算集中趨勢
# 平均值
print(target.mean())

# 中位數
print(target.median()) 

# 眾數
counts = np.bincount(target) # 獲取平鋪後，每個索引位置值，在原始數列中，出現的次數
mode = np.argmax(counts)     # 返回數列中最大值的索引位置
print(counts)
print(mode)
print(target.mode())
```
